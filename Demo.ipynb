```python
import matplotlib.pyplot as plt
import torch.nn

import numpy as np

%matplotlib inline
```

```python
def plot_act(act_fn, mmin=-10, mmax=10, step=0.01):
    x = torch.arange(mmin, mmax, step)
    y = act_fn(x)
    plt.plot(x.numpy(), y.numpy())
    plt.show()
```

# Basic Units

```python
plot_act(torch.nn.ReLU())
```

```python
plot_act(torch.nn.LeakyReLU(negative_slope=0.1, inplace=False))
```

```python
plot_act(torch.nn.LeakyReLU(negative_slope=1, inplace=False))
```

```python
plot_act(torch.nn.Sigmoid())
```

# Tena Units

```python
from tena import *

def plot_group(act_fn, mmin=-10, mmax=10, step=0.01, epochs=200):
    x = torch.arange(mmin, mmax, step)
    rn, cn = epochs // 5 + epochs % 5, 5
    plt.subplots(rn, cn, figsize=(25, 5*rn))
    for ep in range(epochs):
        act_fn.update(ep)
        y = act_fn(x)
        plt.subplot(rn, cn, ep+1)
        plt.title('Epoch: %d' % ep)
        plt.plot(x.numpy(), y.numpy())
    plt.show()
```

## TenaReLU
The same as LeakyReLU.

```python
plot_group(TenaReLU(), epochs=100)
```
